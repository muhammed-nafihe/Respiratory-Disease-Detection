```python
import pandas as pd
audio_features = pd.read_csv("C:/Users/nafih/DM&ML/extracted_audio_features.csv")
patient_diagnosis = pd.read_csv("C:/Users/nafih/DM&ML/Respiratory_Sound_Database/patient_diagnosis.csv")
```

```python
audio_features.head()
```

```python
patient_diagnosis.columns = ['patient_id', 'diagnosis']
patient_diagnosis.head()
```

```python
#Extracting patient id from file name.
audio_features['patient_id'] = audio_features['filename'].str.extract('(\d+)').astype(int)
audio_features['patient_id'].head()
```

```python
#merging both data sets on the basis of feature 'patient_id' and saving to specified folder.
merged_data = pd.merge(audio_features, patient_diagnosis, on='patient_id')
merged_data.to_csv("C:/Users/nafih/DM&ML/merged_dataset.csv", index=False)
```
